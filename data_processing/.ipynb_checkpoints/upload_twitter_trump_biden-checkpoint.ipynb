```python
import pandas as pd
```

```python
#get file object
def get_data_from_text(fileinfo):
    f = open(fileinfo["filename"], "r")
    counter = 0
    date_list = []
    tweet_list = []
    tweet_name_list = []
    while(True):
        line = f.readline()
        counter = counter + 1
        if not line:
            break
#         print("============={}=============================".format(fileinfo['keyword']))
        loc1 =line.find(fileinfo["keyword"])
        date = line[:loc1].split(" ")[1]

        loc2 = line[loc1:].find(" ")
        content = line[loc1:][loc2:]

#     remove https://
        content = content.split('https://')[0].lower()
#         remove @
        at = content.find("@")
        while at>=0:
            space = content[at:].find(" ")
            if space < 0:
                space = content[at:].find(",")
            if space >=0:
                content = content[:at] + content[at + space:]
            else:
                content = content[:at]

            at = content.find("@")
#         Tweet more than 25 words 
        if len(content)>25:
#             print(date, "=====>",content)
            date_list.append(date)
            tweet_list.append(content)
            tweet_name_list.append(fileinfo["keyword"])
    f.close
    return (date_list, tweet_name_list, tweet_list)
```

```python
fileinfo_list = []
fileinfo_list.append({
    'filename': "../data/biden_tweet.txt",
    'keyword': '<JoeBiden>'})
fileinfo_list.append({
    'filename': "../data/trump_tweet.txt",
    'keyword': '<realDonaldTrump>'})
fileinfo_list.append({
    'filename': "../data/nancy.txt",
    'keyword': '<SpeakerPelosi>'})
fileinfo_list.append({
    'filename': "../data/senatemajldr.txt",
    'keyword': '<senatemajldr>'})
date_list_all= []
tweet_list_all = []
tweet_name_list_all = []
for fileinfo in fileinfo_list:
    date_list, tweet_name_list, tweet_list = get_data_from_text(fileinfo)
    date_list_all += date_list
    tweet_list_all += tweet_list
    tweet_name_list_all += tweet_name_list
# @SpeakerPelosi
```

```python
tweetDf= pd.DataFrame({"date": date_list_all, "name": tweet_name_list_all, "tweet": tweet_list_all})
tweetDf
```
